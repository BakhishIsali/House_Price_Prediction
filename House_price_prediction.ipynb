```python
X=df.drop("SalePrice",axis=1)
y=df["SalePrice"]
```

```python
#One Hot Encoding
X = pd.get_dummies(X, columns=cat_cols)
```

```python
X
```

```python
num_columns
```

```python
#Standardizing the Data
from sklearn.preprocessing import StandardScaler
num_columns.remove("SalePrice")
scalar=StandardScaler()
X[num_columns]=scalar.fit_transform(X[num_columns])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
print("X-Train :",X_train.shape)
print("X-Test :",X_test.shape)
print("Y-Train :",y_test.shape)
print("Y-Test :",y_test.shape)
```

```python
#LINEAR REGRESSION
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression()
lin=lin_model.fit(X_train,y_train)
y_pred=lin_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
from sklearn import metrics
```

```python
def evaluation(y,predict):
    mae=mean_absolute_error(y, predict)
    mse=mean_squared_error(y,predict)
    r_squared=r2_score(y,predict)
    return mae,mse,r_squared
```

```python
#MAE : It is the mean of the absolute value of the errors. It measures the difference between two continuous variables, here actual and predicted values of y. 
#^2 : It is a measure of the linear relationship between X and Y. It is interpreted as the proportion of the variance in the dependent variable that is predictable from the independent variable.
#MSE: The mean square error (MSE) is just like the MAE, but squares the difference before summing them all instead of using the absolute value. 
#RMSE: The mean square error (MSE) is just like the MAE, but squares the difference before summing them all instead of using the absolute value. 
mae, mse, r_squared=evaluation(y_test, y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r_squared)
print('RMSE:',np.sqrt(mse))
```
